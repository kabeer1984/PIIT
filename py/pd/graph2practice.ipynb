```python
from plotnine import ggplot,aes
from plotnine import geom_point, geom_col, geom_histogram, geom_boxplot, facet_grid, labs  #imp
```

```python
mtcars = data('mtcars')
mtcars.head(2)
```

```python
ggplot(mtcars) + aes(x="wt", y="mpg", size="hp", color='gear')  + geom_point()
```

```python
ggplot(mtcars) + aes(y="mpg")  + geom_boxplot()
```

```python
ggplot(mtcars) + facet_grid("~ gear") + aes(x='wt',y='mpg', color='am') + labs(x='Weight', y='Mileage') + geom_point()
```

```python
ggplot(mtcars) + aes(x="mpg") + geom_histogram(bins=5, color='green', fill='yellow')
```
