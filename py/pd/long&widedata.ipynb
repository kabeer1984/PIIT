```python
import pandas as pd
import numpy as np
from pydataset import data
```

```python
aq=data('airquality')
```

```python
aq.head(10)
```

```python
aq_long=pd.melt(frame=aq,id_vars=['Month','Day'],value_vars=['Ozone','Solar.R','Wind','Temp'],var_name='measurement',value_name='value')
aq_long.head(10)
```

```python
aq_wide=aq_long.pivot_table(index=['Month','Day'],columns='measurement',values='value').reset_index()
aq_wide.head(10)
```
